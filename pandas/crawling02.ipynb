```python
url = 'https://finance.naver.com/marketindex/exchangeList.nhn'
```

```python
import urllib.request
from bs4 import BeautifulSoup
```

```python
response = urllib.request.urlopen(url)
```

```python
soup = BeautifulSoup(response,'html.parser')
```

```python
soup
```

```python
for result in soup.select("td.tit a"):
    print(type(result.string))
```

```python
for result in soup.select("td.tit a"):
    print(type(result.text))              # 이게 문자열로 바꿔줌
```

```python
result1=[]
for result in soup.select("td.tit a"):
    print(result.text.strip())
    result1.append(result.text.strip())
```

```python
result2 = []
for result in soup.select("td.sale"):
    print(result.text.strip())
    result2.append(result.text)
```

```python
for item in zip(result1,result2):
    print(item)
```

```python
url2 = 'https://music.bugs.co.kr/chart'
```

```python
response2 = urllib.request.urlopen(url2)
```

```python
soup2 = BeautifulSoup(response2,'html.parser')
```

```python
result2_1 = []
for result in soup2.select('p.title'):
    print(result.text.strip())
    result2_1.append(result.text.strip())
```

```python
result2_2 = []
for result in soup2.select('p.artist a:nth-of-type(1)'):  # a태그중 첫번째 것을 뽑겠다.
    print(result.text.strip())
    result2_2.append(result.text.strip())
```

```python
for item in zip(result2_1,result2_2):
    print(item)
```
