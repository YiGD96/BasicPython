```python
from urllib import request,parse
from bs4 import BeautifulSoup
import re
```

```python
url = 'https://news.naver.com/main/read.naver?mode=LSD&mid=shm&sid1=102&oid=011&aid=0003961724'
response = request.urlopen(url)
soup = BeautifulSoup(response,'html.parser')
news = soup.select('ul.list_txt li a')
# print(news)

# 분야별 주요뉴스 가져오기
for n in news:
    title = n['title']
    print('제목 :' ,title)
    
    url = n['href']
    response = request.urlopen(url)
    soup = BeautifulSoup(response,'html.parser')
    content = soup.select_one('#articleBodyContents')
    print(content.text.strip()) # 무슨 내용 있는지 보기
```

```python
list_records=[]
for i in range(1,3):
    params=parse.urlencode({'page':i})
    url="https://movie.naver.com/movie/point/af/list.naver?%s" %params
    response = request.urlopen(url)
    soup = BeautifulSoup(response,'html.parser')
    table = soup.select_one('table.list_netizen')
    for i,tr in enumerate(table.select('tbody tr')):
#         print(i)
#         print(tr)
        for j, td in enumerate(tr.select('td')):
#             print(j)
#             print(td)
            if j==0:
                record = td.string.strip()
                print('글번호 :', record)
            elif j==1:
                record1 = td.select_one('a').string
                print('제목 :',record1)
                record2 = td.select_one('em').string
                print('점수 : ',record2)
                record3 = td.text.replace(record1,'') #이글자 들어가면 빈문자로
                record3 = record3.replace('신고','')
                record3 = re.sub('별점 - 총 10점 중[0-9]{1,2}','',record3).strip() # sub(찾고자하는 패턴[],바꿀문자,여기에서)
                print('감상평 :',record3,'\n')
            record_t = (record,record1,record2,record3)
            list_records.append(record_t)
print(list_records)                
```
